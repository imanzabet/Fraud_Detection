### 3.A.1  Generalized Linear Model (without interactions)

In this model, we build a GLM model with binomial distribution classification using logit. A simple polynomial equation will be used as y = x1 + x2 + ... + xn. We don't see the interactions between variables in this scenario.

```python
poly = PolynomialFeatures(degree=1)
poly_x1_train = poly.fit_transform(x_train)
poly_x1_test = poly.transform(x_test)

clf = sm.GLM(y_train, poly_x1_train, family=sm.families.Binomial(link=sm.families.links.logit()))
res = clf.fit(disp=1)
print(res.summary())
```

```python
from sklearn.metrics import confusion_matrix
y_proba = res.predict(poly_x1_test)
bin_y_proba = [0 if x < 0.5 else 1 for x in y_proba]
print("Confustion Matrix :")
print(confusion_matrix(y_test, bin_y_proba))
print("Classification Report :")
print(classification_report(y_test, bin_y_proba))
```

```python
fpr, tpr, thresholds = roc_curve(y_test,  y_proba)
auc = roc_auc_score(y_test, y_proba)
result_table = result_table.append({'classifiers': 'GLM_Binomial_Poly_1',
                    'fpr':fpr, 
                    'tpr':tpr,
                    'thresholds': thresholds,
                    'auc':auc}, ignore_index=True)
```

```python
# roc curve for logistic regression model with optimal threshold
from numpy import sqrt

fig = plt.figure(figsize=(14,10))

# calculate roc curves
fpr, tpr, thresholds = result_table['fpr'].tolist()[0], \
                        result_table['tpr'].tolist()[0], \
                        result_table['thresholds'].tolist()[0]
# calculate the g-mean for each threshold
gmeans = sqrt(tpr * (1-fpr))
# locate the index of the largest g-mean
ix = np.argmax(gmeans)
print('Best Threshold=%f, G-Mean=%.3f' % (thresholds[ix], gmeans[ix]))
# plot the roc curve for the model
plt.plot([0,1], [0,1], linestyle='-', label='No Skill')
plt.plot(fpr, tpr, linestyle='--', label='Logistic')
plt.scatter(fpr[ix], tpr[ix], marker='o', color='black', label='Best')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
# show the plot
plt.show()
```

We can take a look at the ROC-curve deeper and try to find the optimal "threshold" for the classification.

The best threshold is calculated as 0.49 and is shown on the curve

### 3.A.2  Generalized Linear Model (with interactions terms)

In this model, we build a GLM model with binomial distribution classification using logit. A simple polynomial equation will be used as y = x1 + x2 + ... + xn + (x1*x2) + (x1*x3) ... (xn-1*xn). We do see the interactions between variables in this scenario.

```python
poly2 = PolynomialFeatures(degree=2) #, interaction_only=True, include_bias = False)
poly2_x_train = poly2.fit_transform(x_train)
poly2_x_test = poly2.transform(x_test)

clf = sm.GLM(y_train, poly2_x_train, family=sm.families.Binomial(sm.families.links.logit()))
res = clf.fit(disp=0)
print(res.summary())
```

```python
y_proba = res.predict(poly2_x_test)
bin_y_proba = [0 if x < 0.5 else 1 for x in y_proba]
print("Confustion Matrix :")
print(confusion_matrix(y_test, bin_y_proba))
print("Classification Report :")
print(classification_report(y_test, bin_y_proba))
```

```python
fpr, tpr, thresholds = roc_curve(y_test,  y_proba)
auc = roc_auc_score(y_test, y_proba)
result_table = result_table.append({'classifiers': 'GLM_Binomial_Poly_2',
                    'fpr':fpr, 
                    'tpr':tpr,
                    'thresholds': thresholds,
                    'auc':auc}, ignore_index=True)
```

```python
# Set name of the classifiers as index labels
result_table.set_index('classifiers', inplace=True)

fig = plt.figure(figsize=(14,10))

for i in result_table.index:
    plt.plot(result_table.loc[i]['fpr'], 
             result_table.loc[i]['tpr'], 
             label="{}, AUC={:.3f}".format(i, result_table.loc[i]['auc']))
    
plt.plot([0,1], [0,1], color='orange', linestyle='--')

plt.xticks(np.arange(0.0, 1.1, step=0.1))
plt.xlabel("Flase Positive Rate", fontsize=15)

plt.yticks(np.arange(0.0, 1.1, step=0.1))
plt.ylabel("True Positive Rate", fontsize=15)

plt.title('ROC Curve Analysis', fontweight='bold', fontsize=15)
plt.legend(prop={'size':13}, loc='lower right')

plt.show()
```

### Conclusion
- There is no improvement by adding "interaction" terms to the GLM model

```python
# ML Models Modules
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier



# Turn the values into an array for feeding the classification algorithms.
x_tr = x_train.values
x_ts = x_test.values
y_tr = y_train.values
y_ts = y_test.values
# Let's implement simple classifiers

classifiers = {
    "LogisiticRegression": LogisticRegression(),
    "KNearest": KNeighborsClassifier(),
    "DecisionTreeClassifier": DecisionTreeClassifier(),
    "RandomForestClassifier":RandomForestClassifier(),
    "MLPClassifier":MLPClassifier(),
    "GradientBoostingClassifier":GradientBoostingClassifier()
#     "Support Vector Classifier": SVC()
}
# Wow our scores are getting even high scores even when applying cross validation.
from sklearn.model_selection import cross_val_score


# https://www.imranabdullah.com/2019-06-01/Drawing-multiple-ROC-Curves-in-a-single-plot
# Define a result table as a DataFrame
result_table = pd.DataFrame(columns=['classifiers', 'fpr','tpr','auc'])

for key, classifier in classifiers.items():
    classifier.fit(x_tr, y_tr)
    training_score = cross_val_score(classifier, x_tr, y_tr, cv=5)
    print("Classifiers: ", classifier.__class__.__name__, "Has a training score of", 
                          round(training_score.mean(), 2) * 100, "% accuracy score")
    

    y_proba = classifier.predict_proba(x_ts)[::,1]
    bin_y_proba = [0 if x < 0.5 else 1 for x in y_proba]
    print("Confustion Matrix :")
    cm=confusion_matrix(y_test, bin_y_proba)
    tn, fp, fn, tp = cm[0][0], cm[0][1], cm[1][0], cm[1][1]
    print(cm)
    print(f'Fraud Class precision: {(tn/(tn+fn))*100 :.2f} %')

    fpr, tpr, thresholds = roc_curve(y_test,  y_proba)
    auc = roc_auc_score(y_test, y_proba)

    result_table = result_table.append({'classifiers': classifier.__class__.__name__,
                                        'fpr':fpr, 
                                        'tpr':tpr,
                                        'thresholds': thresholds,
                                        'auc':auc}, ignore_index=True)
    print()
```

#### Conclustion
Among different models, Both Random Forest and Gradient Boosting seem to have promising results.

As we discussed earlier, the accuracy only is not a good metric to evaluate the model. This is due to the fact that the correct classification of "fraud" samples are highly important for classifiers.

The **"Fraude Class Precition"**: (True Negative/(True Negative + False Negative))*100, is introduced to show the precision of the classifier on the Non-Fraud class.

The Multi layer perceptron, shows a terrible accuracy (50%), however it does a great job on classifying the fraud sample with 93.27% precision.

In overall, Random forest classifier, has the best accuracy (75%) and fraud class precision of 76.21%

```python
# Define the outlier detection methods

from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM

classifiers = {
    "Isolation Forest":IsolationForest(n_estimators=100, max_samples=len(x_train), 
                                       contamination=0.5,random_state=random_state, verbose=0),
    "Local Outlier Factor":LocalOutlierFactor(n_neighbors=20, algorithm='auto', 
                                              leaf_size=30, metric='minkowski',
                                              p=2, metric_params=None, contamination=0.5),
    "Support Vector Machine":OneClassSVM(kernel='rbf', degree=3, gamma=0.1,nu=0.05, 
                                         max_iter=-1)
}
```

```python
#Fit the model

n_outliers = Fraud_len
for i, (clf_name,clf) in enumerate(classifiers.items()):
    #Fit the data and tag outliers
    if clf_name == "Local Outlier Factor":
        y_pred = clf.fit_predict(x_test)
        scores_prediction = clf.negative_outlier_factor_
    elif clf_name == "Support Vector Machine":
        clf.fit(x_train)
        y_pred = clf.predict(x_test)
    else:    
        clf.fit(x_train)
        scores_prediction = clf.decision_function(x_train)
        y_pred = clf.predict(x_test)
    #Reshape the prediction values to 0 for Valid transactions , 1 for Fraud transactions
    y_pred[y_pred == 1] = 0
    y_pred[y_pred == -1] = 1
    n_errors = (y_pred != y_ts).sum()
    # Run Classification Metrics
    print("{}: {}".format(clf_name,n_errors))
#     print("Accuracy Score :")
#     print(accuracy_score(y_tr,y_pred))
    print("Confustion Matrix :")
    bin_y_proba = [0 if x < 0.5 else 1 for x in y_pred]
    cm=confusion_matrix(y_test, bin_y_proba)
    tn, fp, fn, tp = cm[0][0], cm[0][1], cm[1][0], cm[1][1]
    print(cm)
    print(f'Fraud Class precision: {(tn/(tn+fn))*100 :.2f} %')
    print("Classification Report :")
    print(classification_report(y_ts,y_pred))
    del y_pred
```

Among "Isolation Forest", "Local Outlier Factor", and "Support Vector Machine" in the unsupervised algorithms, the "Support Vector Machine" shows a better accuracy and precision (50%) on the Non-Fraud class, while it shares the same "Fraud Class precision" (70.26%) with other methods

##### Extended Isolation Forest (Implementation with H2O)

```python
import h2o
from h2o.estimators import H2OExtendedIsolationForestEstimator
h2o.init()

# Import the prostate dataset
h2o_df = h2o.import_file("https://raw.github.com/h2oai/h2o/master/smalldata/logreg/prostate.csv")

# Set the predictors
predictors = ["AGE","RACE","DPROS","DCAPS","PSA","VOL","GLEASON"]

# Define an Extended Isolation forest model
eif = H2OExtendedIsolationForestEstimator(model_id = "eif.hex",
                                          ntrees = 100,
                                          sample_size = 256,
                                          extension_level = len(predictors) - 1)

# Train Extended Isolation Forest
eif.train(x = predictors,
          training_frame = h2o_df)

# Calculate score
eif_result = eif.predict(h2o_df)

# Number in [0, 1] explicitly defined in Equation (1) from Extended Isolation Forest paper
# or in paragraph '2 Isolation and Isolation Trees' of Isolation Forest paper
anomaly_score = eif_result["anomaly_score"]

# Average path length  of the point in Isolation Trees from root to the leaf
mean_length = eif_result["mean_length"]
```

# Conclusions and Future works

Since for this type of classifications, **Type Error II** or in the other word **False Negative** is extremely important. Therefore, metrics such as "accuracy" or "precision" are not very helpful.

The aim of the work is detecting and predicting correct **Fraud** samples. Therefore, metrics such as "recall" or "F1-score" are representing Type II error significantly which is related to the prediction for class 1 or "Fraud" class.

**Notes and Future Works:**
    
    - Dimensionality reduction techniques are also can be used to reduce the amount of features and select the most important ones. There are conventional methods such as PCA as well as deep learning approaches such as Variational Auto Encoder (VAE) with different settings which can help to improve the recall.
    - We can continue model development with other existing model and make a comparison. Such as Baysian models (Gaussian Naive Baysian)
    - There is a potential to use generative statistical models such as Hidden Markov Model (HMM) which is looking to find and generate un-observed (hidden) states and add the new states to the model.
    - There are various "Clustering algorithms" that can be used as unsupervised model to detect the anomalies of the dataset. They can be used individually or mixed of several algorithms to detect the fraud data samples.
    - There is a huge potential for using other boosting models such as "LightGBM" and "Catboost" and their families which are performing very well in categorical datasets.
    - There are "Time-series" approaches that are using deep learning models. These approaches require large amount of memory to craft time-window dataset to train the model. I started to build a LSTM model based on Tensorflow and Keras, however due to technical problem and lack of time, I wasn't able to add this to my current work.
