# 2. Regression Analysis

In this section, the focus will be on **Logistic Regression Analysis**. Defining the dependent variable "isFraud" as a binomial distribution variable, the selected regression analysis would be logistic regression.

Logistic regression works with odds rather than proportions in binomial distributions. To calculate the odds we use this formula: odds=p/(1-p). By transforming the odds in the logarithmic format, we use log(odds) or logit to overcome the infeasibility. We can formulate a logistic regression model as follow:
y=log⁡(p/(1-p))=β_0+β_1 x

## 2.A. Hypothesis testing

In the Multiple logistic regression, there are two hypotheses can be represented. The Null Hypothesis (H0) and Alternative Hypothesis (HA). The H0, assumes all the coefficients in the regression model take zero value, and the HA, assumes the coefficients are not zero. The likelihood ratio test based on the Chi-square distribution calculates the significance of the difference between the likelihood ratio of the model with non-zero predictors and the model with zero predictors (only intercept). Usually, the significance level (α=0.05). 

	H0: There is no relationship between the logit of the dependent variable and independent variables or β_1=β_2=⋯=β_n=0 (β_j for j=1..n is not significant)
    
	HA: There is a relationship between the logit of the dependent variable and independent variables or β_k≠0 for at least one k (β_k is significant)

## Build Logistic Regression Model for Hypothesis Testing

In the logistic regression model, we are looking to estimate the regression parameters and define the fitted model equation. We also look to identify the 95% confidence interval for both regression slope and intercept. In the hypotheses testing, defining significance test results for the zero slope in the null hypothesis is important. And finally, we would interpret the odds ratio along with its 95% confidence interval. In general, statistical inference in logistic regression is similar to linear regression. The difference is that we use standard Normal z*-values instead of using critical values from the t-distributions. 

Therefore the basis of the hypotheses tests is “the ratio of the estimated slope to the standard error”. The square of this ratio is often used in statistical tests. Hence, the P-values are obtained from the chi-square (X^2) distribution with DoF 1.


we build multiple logistic model. To train the model, we split the data into 60% train and 40% test data.

```python
from sklearn.model_selection import train_test_split
col_names = df.columns.tolist()

LIMIT = -1
dfTemp = df[:LIMIT]
x_train,x_test,y_train,y_test = train_test_split(dfTemp[[c for c in col_names if c not in ['isFraud']]],
                                                 dfTemp[['isFraud']],test_size = 0.2,random_state = 22) 
```

```python
# importing libraries
import statsmodels.api as sm
import numpy as np


x_train = sm.add_constant(x_train)
x_test = sm.add_constant(x_test)

# Adding noise to avoid singularity error (inverse matrix calculation)
# n = 1000000
x_trainDirty = x_train+0.0000001*np.random.rand(x_train.shape[0], x_train.shape[1])
x_testDirty = x_test+0.0000001*np.random.rand(x_test.shape[0], x_test.shape[1])

# building the model and fitting the data
log_reg = sm.Logit(y_train, x_trainDirty).fit()
log_reg.summary()
```

>We can interpret that p-values are significant (p>0.05) for some of the variables. Therefore we failed to reject the null hypothesis, and we conclude there is no association of these variables with the dependent/output variable ("isFraud"). 

>However the small p-values (p<0.05) for other independent variables (CVV_diff, cardPresent, merchantCategoryCode, posConditionCode, posEntryMode, transactionAmount, customerId, and index), reject the null hypothesis and shows that there is a significant relationship between them and the dependent variable.

This approach is useful since it gives us an idea about the significant variables in the logistic regression analysis. Defining the significant variables helps to eliminate the non-significant variables for future analysis and building machine learning models.

## 2.B Logistic Regression Statistical Assumptions

In Logistic Regression analysis, there are a few statistical assumptions where need to be examined before further analysis.
   >1- Residual Normality Check: Since the dependent variable is binary in logistic regression, the residuals (differences between observed and predicted values) from the logistic regression model need to be transformed. The residuals can be assessed with one of these methods: Pearson residuals, Studentized Pearson residuals, or deviance residuals. In this work, Deviance Residuals and Studentized Pearson Residuals against Fitted Values will be examined.
   
   >2- Multicollinearity: When the data contain highly correlated predictor/independent variables, multicollinearity is happening. When the predictors are highly correlated, it causes a problem for solving the linear algebraic form of the regression equation. In this work Variance Inflation Factor (VIF) is used to detect and remove multicolliniearity.
   
   >3- Sample size: In general, logistic regression works best with a large sample size of the dataset. The sample size needs to be more than 25 or 30 samples.
   
   >4- Features are Directional: Logistic regression assumes that the relationship between features is unidirectional. In other word, there are not multiple (two or more) peaks (multimodality) in the scatter plots since the slope represents one value β.
   
   >5- Heteroscedasticity Check: Is not required
   
   >6- Normality Check: Is not required.
   
   >7- Linearity: Is not required. 

#### 2.B.1 Residual Normality Check

```python
reg.residual_plot(log_reg)
```

Since the dependent variable is binary in logistic regression, the residuals (differences between observed and predicted values) from the logistic regression model need to be transformed. We know that the nature of the outcome variable is binary. Therefore the residuals will be unknown (not normally distributed) (Kutner, 2005)
The residuals can be assessed with one of these methods: Pearson residuals, Studentized Pearson residuals, or deviance residuals. Using the deviance residuals method, we would estimate the probability or linear predictors along with using Lowess smooth. In the case of a perfect logistic regression model, we know that the residuals (differences between the observed and predicted values) are equal to zero. (Kutner, 2005) suggests that using lowess smooth the plot for each of Studentized Pearson residuals or Deviance residuals would approximately be a zero-intercept horizontal line. 

In the residual normality check analysis (the above diagrams), the distribution of residual of the fitted values of the model for the Deviance Residuals (first plot) and the Studentized Pearson Residuals (second plot) is shown.

In the first plot, the distribution of residual is almost following a normal distribution. And for the Studentized Pearson Residuals (second plot) the distribution is a following exactly the normal distribution.

Therefore, we conclude and confirm the residual normality check assumption.

REFERENCE:

Kutner, M. H. (2005). Applied linear statistical models.

#### 2.B.2 Multicollinearity check (VIF approach)

The methods to detect multicollinearities are Bunch Map analysis, correlation methods, or Variance Inflation Factor ($VIF$). We fit a model based on the entire input variables and then calculate VIF using the $\frac{1}{1-R^2}$ the equation for each variable. Where $R^2$, is the R-squared for regression. VIF simply shows how much the variance of the coefficient can increase when the input variables are correlated. The higher VIF shows the higher correlation to other variables. We defined the threshold of 5 for VIF, tried to eliminate the features with $VIF>5$ in each iteration, and re-calculated the VIF for the new features. 

```python
x = df[[c for c in col_names if c not in ['isFraud']]]
out = reg.calculate_vif(x) 
# out.head()
```

```python
x = df[[c for c in col_names if c not in ['isFraud', 'merchantCountryCode', 'availableMoney', 'acqCountry',
                                        'transactionType', 'currentBalance', 'merchantCategoryCode']]]
out = reg.calculate_vif(x) 
out.head()
```

```python
# showing the final list
print(reg.calculate_vif2(x))
```

To examine the **multicollinearity**, the Variance Inflation Factor (VIF) approach is used. The threshold has been set for around 5, i.e. when the amount of VIF for each dependent variable exceeds the threshold, the variable is removed in each iteration.

## 2.D Statistical Conclusion Validity

   1- **Sample Size**: Using a conservative value for Cohen’s d=0.8 (threshold) of the effect size, we hope to get the minimum of 80% power in our statistical conclusion. Based on these assumptions, we can calculate the minimum samples size required for our analysis of 26.
    
   2- **Error Type I and Error Type II**: Type I Error happens when the null hypothesis is rejected; while it shouldn’t be rejected in reality, the null hypothesis was confirmed. Type I Error is denoted as α, showing the significance level or the false negative. Type II Error happens when the null hypothesis is accepted, while the null hypothesis is false in reality. And denoted as β and referred to false positive. 
    
   3- **Effect Size**: It quantifies the effect of the experiment on the population. Effect size has two main groups of methods association-based and difference-based. One of the popular association-based methods is Pearson’s correlation method. By calculating the Pearson’s correlation coefficient, we understand about positive/negative impact on the relationship between variables.
    
   4- **Statistical Power**: Based on the calculations of power of independent samples based on F-test ANOVA, we can plot the relationships of power and sample size for different effect sizes. The figure below shows that to get an acceptable power for effect size of d=0.8, we need sample sizes greater than 25, which confirmed our previous calculation of sample size above.
